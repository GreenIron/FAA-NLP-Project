```python
import pandas as pd
import pickle
import glob
import os
import re

ata2D = re.compile("^[0-9]{2}")
redots = re.compile("\.{2,30}")

techpub = [[]]
for mypickle in glob.glob(r"C:\Users\victor\Documents\DeepLearning\FAA NLP Project\database\ATA\ORION\*.pickle"):
    e = []
    try:
        with open(mypickle,"rb") as f:
            pages = pickle.load(f)
    except:
        e = 'failed'

    if e==[]:        
        print(mypickle)
        for pages in pages:
            if ata2D.search(pages[0][-1]):
                techpub.append([os.path.basename(mypickle),
                                os.path.basename(mypickle).split('__')[1],
                                '. '.join(pages[0]),
                                pages[0][-1][0:2],
                                redots.sub('.', '\n\n '.join(pages[2:]).replace('\n', ' ').replace('..', '.'))                                
                                ])
        
df_techpub = pd.DataFrame(techpub,
                            columns=['source', 
                                    'doc type'
                                    'title', 
                                    'ATA2D',
                                    'raw text'])    
df_techpub = df_techpub.astype("string")

df_techpub.to_parquet(r'C:\Users\victor\Documents\DeepLearning\FAA NLP Project\database\orion.parquet')
```
